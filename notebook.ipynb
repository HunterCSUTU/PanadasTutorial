```python
dirty_data = {
    'Student': ['Emma', 'Liam', None, 'Noah'],
    'Age': [20, None, 19, 22],
    'Major': ['Math', 'Physics', 'Biology', None],
    'Grade': [88, 92, None, 95]
}

dirty_df = pd.DataFrame(dirty_data)
dirty_df
```

```python
dirty_df['Age'] = dirty_df['Age'].fillna(0)
dirty_df['Major'] = dirty_df['Major'].fillna("Unknown")
dirty_df = dirty_df.rename(columns={'Grade':'Score'})
dirty_df['Score'] = dirty_df['Score'].astype(float)
dirty_df
```

groupby() in pandas is one of the most important tools for data analysis. It lets you split your data into groups, apply a calculation to each group, and combine the results back together.

Think of it like:

“Put similar rows together, then run math on each group separately.”

📊 What groupby() does

When you do:

df.groupby('Major')

pandas:

Splits the DataFrame into groups based on unique values in "Major"
Creates separate mini-DataFrames for each group
Lets you apply functions to each group

🧠 Basic structure

df.groupby('column')[target_column].function()

```python
df.groupby('Major').size()
```

```python
dirty_df.groupby('Major')['Score'].mean()
```
